```python
import matplotlib.pyplot as plt
import csv
```

```python
x = []
y = []

def plot_cpu(pid):
    with open(f"./logs/{pid}.csv",'r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        for row in lines:
            x.append(row[0])
            y.append(float(row[1]))

    plt.plot(x, y, color = 'g', linestyle = 'dashed',
            marker = 'o',label = "CPU Usage")
    plt.xticks(rotation = 90)
    plt.xlabel('Time')
    plt.ylabel('CPU Usage Percent')
    plt.title('CPU Usage', fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
xx=[]
yy=[]

def plot_ram(pid):
    with open(f"./logs/{pid}.csv",'r') as csvfile:
        lines = csv.reader(csvfile, delimiter=',')
        for row in lines:
            xx.append(row[0])
            yy.append(float(row[2]))
    plt.plot(xx, yy, color = 'g', linestyle = 'dashed',
            marker = 'o',label = "RAM Usage")
    plt.xticks(rotation = 90)
    plt.xlabel('Time')
    plt.ylabel('RAM Usage Percent')
    plt.title('RAM Usage', fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()
```

```python
def plot(pid):
    plot_cpu(pid)
    plot_ram(pid)
```

```python
plot(31380)
```
